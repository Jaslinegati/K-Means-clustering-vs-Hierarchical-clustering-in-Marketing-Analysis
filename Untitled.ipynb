## Modeling





### K-MEANS

Since K-means is an unsupervised-learning type, the class attribute is not necessary for algorithm execution. In the section ' Understanding the Meaning' we stated that the attribute ' Revenue' will be used as the mark of class. It will be extracted in another variable, and stored. We can then normalize the attributes left behind.

```r
# creating a new dataframe
Advert.new <- Advert[,c(1:17)]
colnames(Advert.new)
```

```r
# storing the class attribute in another variable
Advert.class <- Advert["revenue"]
colnames(Advert.class)
```

```r
# Normalising the data so that no particular attribute has more impact on clustering than others
normalize <- function(x){
return ((x-min(x)) / (max(x)-min(x)))
}
Advert.new$administrative <- normalize(Advert.new$administrative)
Advert.new$administrative_duration <- normalize(Advert.new$administrative_duration)
Advert.new$infromational <- normalize(Advert.new$informational)
Advert.new$infromational_duration <- normalize(Advert.new$informational_duration)
Advert.new$productrelated <- normalize(Advert.new$productrelated)
Advert.new$productrelated_duration <- normalize(Advert.new$productrelated_duration)
Advert.new$bouncerates<- normalize(Advert.new$bouncerates)
Advert.new$exitrates<- normalize(Advert.new$exitrates)
Advert.new$pagevalues<- normalize(Advert.new$pagevalues)
Advert.new$specialday<- normalize(Advert.new$specialday)
Advert.new$opertaingsystem<- normalize(Advert.new$operatingsystems)
Advert.new$browser<- normalize(Advert.new$browser)
Advert.new$region<- normalize(Advert.new$region)
Advert.new$traffictype<- normalize(Advert.new$traffictype)
head(Advert)
```

```r
names(Advert.new)
```

```r
# Storing numeric values in a new variable
Advert.clust <- Advert.new[,c('administrative','administrative_duration','informational','informational_duration','productrelated', 
                              'productrelated_duration','bouncerates','exitrates','pagevalues','specialday','operatingsystems' 
                              ,'browser','region','traffictype' )]
# Apply k means clustering algorithm with a number of centroids k
result<- kmeans(Advert.clust,3)
```

```r
# preview the number of records in each cluster
result$size
```

```r
# get the value of cluster center datapoint value(k centers for k)
result$centers
```

```r
# getting cluster vector that shows the cluster where each record falls
result$cluster
```

### Hierachical Clustering

```r
# Compute descriptive stats
desc_stats <- data.frame(
Min = apply(Advert.clust, 2, min), # minimum
Med = apply(Advert.clust, 2, mean), # mean
SD = apply(Advert.clust, 2, sd), # Standard deviation
Max = apply(Advert.clust, 2, max) # Maximum
)
desc_stats <- round(desc_stats, 1)
head(desc_stats)
```

```r
# scale the datafarm to make the variables comparable
# entails transforming variables such that they have a mean of 0 and std dev of 1
# because we dont want the hierachical clustering result to depend on an arbitrary variable unit
Advert.clust <- scale(Advert.clust)
head(Advert.clust)
```

```r
# clustering
# use the R function hclust() for hierarchical clustering
# 1st use dist() function to compute the Euclidean Distance btwn observs
# d will be the 1st argument in the hclust() function dissimilarity matrix
d<-dist(Advert.clust,method = "euclidean")
res.hc <- hclust (d,method = "ward.D2")
```

```r
# plot obtained dendogram
plot(res.hc,cex = 0.6,hang = -1)
```

```r
# checking for accuracy
mean(Advert.clust== result$cluster)
```
